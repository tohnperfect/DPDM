## Load Data

```
from sklearn import datasets

iris_data = datasets.load_iris()
X = iris_data['data']
y = iris_data['target']
```

# Decision Tree

CaptureD3.JPG

## Import

```
from sklearn.tree import DecisionTreeClassifier
```

## Define

```
myTree = DecisionTreeClassifier(criterion='entropy',max_depth=2)
```

## Train

```
myTree.fit(X[:-2,:],y[:-2])
```

## Test

```
myTree.predict(X[-2:,:])
```

## Draw Tree

```
from sklearn.tree import plot_tree
```

```
plot_tree(myTree);
```
